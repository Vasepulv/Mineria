```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.dummy import DummyClassifier
from sklearn.svm import SVC  # support vector machine classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB  # naive bayes
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.neighbors.classification import KNeighborsClassifier

db = pd.read_csv("DEF_2017.csv", sep=";", low_memory=False) # abrimos el archivo csv y lo cargamos en data.

db = db[["EDAD_CANT", "SEXO", "NIVEL_INS", "URBA_RURAL", "REG_RES", "COMUNA", "DIAG1", "OCUPA", "ACTIV", "CATEG"]]
db = db.drop(["COMUNA"], axis=1)
db = db.drop(["CATEG"], axis=1)

db["EDAD_CANT"] = db["EDAD_CANT"] // 20
db["SEXO"] = db["SEXO"] - 1
db["URBA_RURAL"] = db["URBA_RURAL"] - 1
db["DIAG1"] = db["DIAG1"].str.slice(0, 2)
aux=db["DIAG1"].str.contains('C',na=False, regex=True)
db=db[aux]

aux2=db["SEXO"]==0
db=db[aux2]

db["DIAG1"]=db["DIAG1"].replace(["C0","C5","C7","C8","C4","C9"],"Otros")
#db["DIAG1"]=db["DIAG1"].replace(["A", "B", "D", "E", "F", "G", "H", "L", "M", "N", "O"], "Otros",regex=True)

db['ACTOCU'] = db["ACTIV"].map(str)+ "-" + db["OCUPA"].map(str)
db = db.drop(["ACTIV"], axis=1)
db = db.drop(["OCUPA"], axis=1)

db = pd.concat([db, pd.get_dummies(db['EDAD_CANT'], prefix='Grupo Etario')], axis=1)
db = db.drop(['EDAD_CANT'],axis=1)

db = pd.concat([db, pd.get_dummies(db['NIVEL_INS'], prefix='Nivel Educacion')], axis=1)
db = db.drop(['NIVEL_INS'],axis=1)

db = pd.concat([db, pd.get_dummies(db['REG_RES'], prefix='Region')], axis=1)
db = db.drop(['REG_RES'],axis=1)

db = pd.concat([db, pd.get_dummies(db['ACTOCU'], prefix='Actividad+Ocupacion')], axis=1)
db = db.drop(['ACTOCU'],axis=1)

dbx = db.drop(columns=["DIAG1"])

print(dbx.iloc[:5, :])

from sklearn.neighbors.nearest_centroid import NearestCentroid
clf = NearestCentroid()
#clf = DummyClassifier()
X_train, X_test, y_train, y_test = train_test_split(dbx, db['DIAG1'], test_size=0.3,stratify=db['DIAG1'])
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test) 
print("Accuracy en test set:", accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```
